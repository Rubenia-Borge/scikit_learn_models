Plot histograms for the numerical features

```python
# Write your code here.
_ = penguin_census.hist(figsize=(20, 14))
```

Show features distribution for each class. Hint: use
[`seaborn.pairplot`](https://seaborn.pydata.org/generated/seaborn.pairplot.html)

```python
# Write your code here.
import seaborn as sns

# We will plot a subset of the data to keep the plot readable and make the
# plotting faster
n_samples_to_plot = 20
target_column = "Species"
columns = ["Culmen Length (mm)", "Culmen Depth (mm)"]
_ = sns.pairplot(
    data=penguin_census[:n_samples_to_plot],
    vars=columns,
    hue=target_column,
    plot_kws={"alpha": 0.2},
    height=3,
    diag_kind="hist",
    diag_kws={"bins": 30},
)
```
